```python
# 981
mod = 10**9+7

n = 10**5+1
fact = {}

def computeFact():

    global fact

    ans = 1
 

    # Iterate in the range [1, n]

    for i in range(1,n+1,1):

        # Update ans to ans*i

        ans = (ans * i) % mod
       # Store the value of ans

        # in fact[i]

        fact[i] = ans

 
# Function to check whether the
# current character is a vowel or not

def checkVowles(v):

    if (v == 'A' or v == 'E' or v == 'I' or v == 'O' or v == 'U'):

        return True

    else:

        return False
 
# Function to count the number of
# new words of S satisfying the
# given condition

def countWords(s):

    # Store the facttorials upto n

    global fact
 

    

    # all facttorials upto n

    computeFact()
 

    # frequencies of all characters

    count = {}
 

    # vowels and consonants

    vowels = 0

    cons = 0
 

    # Iterate through all

    # characters in the string

    for i in range(len(s)):

        # Update the frequency

        # of current character

        if s[i] in count:

            count[s[i]] += 1

        else:

            count[s[i]] = 1
 

        # Check if the character

        # is vowel or consonant

        if (checkVowles(s[i])):

            vowels += 1

        else:

            cons += 1
 

    # Check if sun(cons)  == sum(vowels+1) or sum(cons) == sum(vowles)

    if ((cons == vowels + 1) or (cons == vowels)):

        # Store the denominator

        deno = 1
 

        # Calculate the denominator

        # of the expression

        for key,value in count.items():

            # Multiply denominator by facttorial

            # of counts of all letters

            deno = (deno * fact[value]) % mod
 

        # Store the numerator

        nume = fact[cons] % mod

        nume = (nume * fact[vowels]) % mod
 

        # Store the answer by dividing

        ans = nume // deno
 

        # Print the answer

        return ans


    else:

       return 0
# Driver code:
countWords("CAT")
```
