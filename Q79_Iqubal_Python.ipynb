```python
fileName = input()

with open(fileName) as file:
  numofEmp = 0 # to store the count of the employees whose salary is more the 50k
  sumofSalary = 0

  for l in file:
    dataLine = l.split()
    print(dataLine)
    salary = int(dataLine[len(dataLine)-1])

    # check salary >= 50k
    if (salary>50000):
      numofEmp +=1
      sumofSalary += salary

print(numofEmp)
print(sumofSalary)

file = open("new_"+fileName,"w")
file.write(str(numofEmp) + "\n" + str(sumofSalary) + "\n")
file.close()
```
